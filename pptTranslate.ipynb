````python
# （勿刪）這邊利用Drive API將pptx存成google slides

def upload_and_convert_to_slides():
    """上傳 .pptx 檔案並轉換為 Google Slides。

    Returns:
      str: 上傳的 Google Slides 檔案的 ID。
    """

    creds, _ = google.auth.default()

    try:
        # 建立 Drive API 客戶端
        service = build('drive', 'v3', credentials=creds)

        # 設定檔案 metadata 和 media content
        file_metadata = {'name': 'slides_71-87',
                         'mimeType': 'application/vnd.google-apps.presentation'}  # 設定 Google Slides 檔案的名稱
        media = MediaFileUpload(
            '/content/drive/MyDrive/App/translate/slides_71-87.pptx',  # 替換成妳的 .pptx 檔案路徑
            mimetype='application/vnd.openxmlformats-officedocument.presentationml.presentation',
            resumable=True,  # 啟用 resumable upload，適用於超過 5MB 的檔案

        )

        # 建立 Google Slides 檔案
        file = service.files().create(
            body=file_metadata,
            media_body=media,
            fields='id'
        ).execute()

        # 將檔案轉換為 Google Slides 格式
        file_id = file.get('id')
        mime_type = 'application/vnd.google-apps.presentation'
        file = service.files().update(
            fileId=file_id,
            body={'mimeType': mime_type},
            fields='id'
        ).execute()

        print(f'已建立 Google Slides 檔案：{file.get("id")}')
        return file.get('id')

    except HttpError as error:
        print(f'發生錯誤：{error}')
        return None

# if __name__ == '__main__':
#     upload_and_convert_to_slides()
````

````python
import re

def clean_api_response_string(s: str) -> str:
    """
    【簡化穩定版】清理 API 回傳的字串，只處理最常見的核心問題。
    """
    if not isinstance(s, str) or not s.strip():
        return ""

    # 步驟 1: 移除頭尾可能存在的 Markdown 區塊符號和空白
    cleaned_s = s.strip(" `\n")
    if cleaned_s.startswith("json"):
        cleaned_s = cleaned_s[4:].strip() # 移除 json 和之後的空白

    # 步驟 2: 使用正則表達式，安全地替換所有在引號之外的全形逗號 '，'
    # 這是處理逗號最穩健的方法
    try:
        # 這個正則表達式會尋找不在 ('...' 或 "...") 內的 '，'
        cleaned_s = re.sub(r"""，(?=(?:[^"']|"[^"]*"|'[^']*')*$)""", ",", cleaned_s)
    except Exception as e:
        print(f"清理全形逗號時發生正則表達式錯誤: {e}")
        # 如果正則表達式失敗，退回使用簡單的全局替換作為備案
        cleaned_s = s.replace('，', ',')

    return cleaned_s
````

````python
import re
from google.api_core.exceptions import PermissionDenied, ResourceExhausted, InvalidArgument
import google.generativeai as genai

# 這裡假設你已經完成了 genai.configure(api_key=YOUR_API_KEY)
# 如果沒有，請在調用此函數前執行

def translateWithGemini(combined_text):
    """
    呼叫 Gemini API 將合併後的英文文字翻譯成中文。

    Args:
        combined_text (str): 待翻譯的完整英文字串。

    Returns:
        str: 翻譯後的繁體中文字串，若翻譯失敗則返回原始字串。
    """

    model = genai.GenerativeModel("gemini-2.5-flash-lite")

    try:
        response = model.generate_content(
            f'''
            請將以下文字翻譯成台灣繁體中文。
            文字中會穿插類似@@@TRANSLATION_SPLIT@@@的間隔符號，你必須在翻譯時考慮整段文字合起來的完整語意，而不受間隔符號所影響。
            並且，因為中文和英文的語序本來就不一樣，請不必拘泥原來英文的語序，以最通順的方式呈現中文，但是返回給我的翻譯必須將間隔符號加回合適的地方。
            原來有幾個間隔符號，就要加回去幾個，如果原本有\n也必須加回去。
            原始文字：
            {combined_text}
            '''
        )
        # 清除 API 回傳中可能包含的額外內容，只保留翻譯的純文字
        translated_text = response.text.strip().strip('```').strip()
        return translated_text

    except PermissionDenied:
        print("API 金鑰無效或沒有權限存取 Gemini API。")
        return combined_text

    except ResourceExhausted:
        print("超過 Gemini API 的使用配額。")
        return combined_text

    except InvalidArgument as e:
        print(f"無效的參數：{e}")
        return combined_text

    except Exception as e:
        print(f"呼叫 Gemini API 發生錯誤：{e}")
        return combined_text
````

````python
import os
import re
from googleapiclient.errors import HttpError

# 這裡假設 slides_service 和 presentation 都已正確設定
# 假設你已經將上面我給你的新 translateWithGemini 函數定義在程式碼中

forbidden_string = 'Evaluation only.Created with Aspose.Slides for Python via .NET 25.5.Copyright 2004-2025Aspose Pty Ltd.'

for slide in presentation['slides'][39:40]:
    slides_id = slide['objectId']
    print(f"\n==================== 正在處理頁面 {slides_id} ====================")

    deleted_object_ids = set()
    translation_log = [] # 新增：用於記錄所有翻譯日誌

    # 唯一佔位符，用於分隔不同 textRun 的內容
    # 這裡使用一個不太可能出現在簡報中的字串作為佔位符
    PLACEHOLDER = "@@@TRANSLATION_SPLIT@@@"

    # ======================================================================
    # 階段一：檢查並立即刪除 Forbidden String 元素
    # （此部分與原文相同，故保留，不進行修改）
    # ======================================================================
    print("\n--- 階段一：正在檢查並刪除 Forbidden String 元素 ---")
    for page_element in list(slide['pageElements']):

        def check_and_delete_if_forbidden(shape_element):
            """檢查 Shape 文字，如果符合 forbidden_string 就立即刪除。"""
            if 'text' in shape_element.get('shape', {}) and 'textElements' in shape_element['shape']['text']:
                full_text = ''.join(
                    te.get('textRun', {}).get('content', '')
                    for te in shape_element['shape']['text']['textElements']
                )
                comparable_text = full_text.replace('\n', '')

                if comparable_text == forbidden_string:
                    object_id_to_delete = shape_element['objectId']
                    print(f"發現 Forbidden String 於元素 {object_id_to_delete}，正在立即刪除...")
                    try:
                        requests = [{'deleteObject': {'objectId': object_id_to_delete}}]
                        body = {"requests": requests}
                        slides_service.presentations().batchUpdate(presentationId=presentation_id, body=body).execute()
                        print(f"元素 {object_id_to_delete} 已成功刪除。")
                        deleted_object_ids.add(object_id_to_delete)
                        return True
                    except HttpError as error:
                        print(f"!!! 刪除元素 {object_id_to_delete} 時發生錯誤: {error}")
            return False

        if 'shape' in page_element and 'elementGroup' not in page_element.get('description', ''):
            check_and_delete_if_forbidden(page_element)
        elif 'elementGroup' in page_element:
             for child in page_element['elementGroup'].get('children', []):
                if 'shape' in child:
                    check_and_delete_if_forbidden(child)

    # ======================================================================
    # 階段二：收集、翻譯與替換剩下的正常文字
    # ======================================================================
    print("\n--- 階段二：正在處理剩餘元素的翻譯與替換 ---")

    # 1. 收集所有需要翻譯的文字
    combined_original_texts = []
    text_elements_to_process = []

    def collect_text(elements):
        for element in elements:
            if 'textRun' in element and 'content' in element['textRun'] and any(char.isalnum() for char in element['textRun']['content']):
                original_text = element['textRun']['content']
                combined_original_texts.append(original_text)
                text_elements_to_process.append(element)

    for page_element in slide['pageElements']:
        if page_element['objectId'] in deleted_object_ids:
            continue

        if 'shape' in page_element and 'text' in page_element['shape']:
            collect_text(page_element['shape']['text']['textElements'])
        elif 'table' in page_element:
            for row in page_element['table'].get('tableRows', []):
                for cell in row.get('tableCells', []):
                    if 'text' in cell:
                        collect_text(cell['text']['textElements'])
        elif 'elementGroup' in page_element:
            for child in page_element['elementGroup'].get('children', []):
                if child['objectId'] in deleted_object_ids:
                    continue
                if 'shape' in child and 'text' in child['shape']:
                    collect_text(child['shape']['text']['textElements'])

    # 2. 合併文字並進行翻譯
    if combined_original_texts:
        combined_text = PLACEHOLDER.join(combined_original_texts)
        print(f"\n>>> 正在將 {len(combined_original_texts)} 個文字片段合併並發送翻譯請求...")

        try:
            os.environ['HTTP_PROXY'] = ''
            os.environ['HTTPS_PROXY'] = ''
            translated_text = translateWithGemini(combined_text)

            # 將完整的翻譯字串按佔位符分割
            translated_parts = translated_text.split(PLACEHOLDER)

            if len(translated_parts) != len(combined_original_texts):
                print("!!! 警告：翻譯後的文字片段數量與原始數量不匹配。")

            # 3. 逐一替換並記錄日誌
            requests_for_replacement = []
            for i, original_text in enumerate(combined_original_texts):
                translated_replacement = translated_parts[i] if i < len(translated_parts) else original_text

                # 記錄翻譯日誌
                translation_log.append({
                    'original': original_text,
                    'translated': translated_replacement
                })

                requests_for_replacement.append({
                    'replaceAllText': {
                        'pageObjectIds': [slides_id],
                        'containsText': {'text': original_text, 'matchCase': False},
                        'replaceText': translated_replacement
                    }
                })

            # 4. 執行批次更新
            if requests_for_replacement:
                try:
                    body = {"requests": requests_for_replacement}
                    print(f"\n>>> 正在對頁面 {slides_id} 執行 {len(requests_for_replacement)} 項文字替換...")
                    slides_service.presentations().batchUpdate(presentationId=presentation_id, body=body).execute()
                    print(">>> 文字替換完成！")
                except Exception as e:
                    print(f"!!! 執行批次更新時發生錯誤: {e}")

        except Exception as e:
            print(f"!!! 翻譯或處理文字時發生錯誤: {e}")

    print("\n--- 翻譯日誌 ---")
    for log_entry in translation_log:
        print(f"原文: {log_entry['original'].replace('\n', '\\n')}")
        print(f"譯文: {log_entry['translated'].replace('\n', '\\n')}")
        print("-" * 20)
````
